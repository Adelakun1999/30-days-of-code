```python
X=df.drop('class',axis=1)
```

```python
y=df['class']
```

```python
X.head()
```

```python
X=pd.get_dummies(X,drop_first=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
help(AdaBoostClassifier)
```

```python
model=AdaBoostClassifier(n_estimators=1)
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report, plot_confusion_matrix
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
print(classification_report(y_test,predictions))
```

```python
model.feature_importances_
```

```python
model.feature_importances_.argmax()
```

```python
sns.countplot(x='odor',data=df, hue='class')
```

```python
error_rates=[]

for n in range(1,96):
    model=AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    
    err=1 - accuracy_score(y_test,preds)
    
    error_rates.append(err)
```

```python
plt.plot(range(1,96),error_rates)
```

```python
model.feature_importances_
```

```python
feat=pd.DataFrame(model.feature_importances_, columns=['Importance'], index=X.columns)
```

```python
feat
```

```python
imp_features=feat[feat['Importance']>0].sort_values('Importance')
```

```python
imp_features
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.barplot(data=imp_features, x=imp_features.index, y='Importance')
plt.xticks(rotation=90);
```

```python
print(classification_report(y_test,model.predict(X_test)))
```
